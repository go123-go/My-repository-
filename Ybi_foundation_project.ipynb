**Pandas Library usage:**
Data cleaning and preprocessing to handle the missing value.df.resample() for time based analysis ( eg. daily or monthly average )and df.plot() for creating various plot types.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# sample air qualities data from Indian cities
data={
    'city':['Delhi','Mumbai','Kolkata','Chennai','Bengaluru','Hyderabad','Pune','Ahmadabad','Jaipur','Lucknow'],
    'date':pd.to_datetime(['2023-01-15','2023-01-15','2023-01-15','2023-01-15','2023-01-15','2023-01-15','2023-01-15','2023-01-15','2023-01-15','2023-01-15']),
    'pm25':[250,120,180,90,70,150,110,160,200,140],
    'no3':[80,50,60,40,30,70,45,65,90,55],
    'so2':[30,20,25,15,10,25,18,28,35,22],
}
air_quality_df=pd.DataFrame(data)
# 1. Basic data exploration
print(air_quality_df.head())
print(air_quality_df.info())
```

```python
# 2.PM2.5 Analysis
average_pm25=air_quality_df['pm25'].mean()
print("\n Average PM2.5 across cities:", average_pm25)
```

```python
# 3.City with highest NO3 levels
city_highest_no3=air_quality_df.loc[air_quality_df['no3'].idxmax(),'city']
print ("\n City with highest NO3 levels:",city_highest_no3)
```

```python
# 4.Correlation between pollutants
correlation_pm25_no3=air_quality_df['pm25'].corr(air_quality_df['no3'])
print ("\n correlation between PM2.5 and NO3:",correlation_pm25_no3)
```

```python
# 1.Bar chart of PM2.5 levels by city
plt.figure(figsize=(12,6))
sns.barplot(x="city",y="pm25",data=air_quality_df)
plt.title('PM2.5 Levels by city')
plt.xlabel('city')
plt.ylabel('pm2.5')
plt.xticks(rotation=45)
plt.show()
```

```python
 #2. Scatter plot of pm2.5 vs. no3
plt.figure(figsize=(8,5))
sns.scatterplot(x='pm25',y='no3',data=air_quality_df,hue='city')
plt.title('pm2.5 vs.no3')
plt.xlabel('PM2.5')
plt.ylabel('NO3')
plt.show()
```
